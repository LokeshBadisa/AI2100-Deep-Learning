```python
#Generating Gaussian Kernel
def gkern(l=3, sig=1e-5):
    ax = np.linspace(-(l - 1) / 2., (l - 1) / 2., l)
    gauss = np.exp(-0.5 * np.square(ax) / np.square(sig))
    kernel = np.outer(gauss, gauss)
    return kernel / np.sum(kernel)
```

```python
#Generating Mean Kernel
def mkern(l=3):
    l=float(l)
    return np.ones((int(l),int(l)))*(-1/(l)**2)
```

```python
#Helper function for checkimgkersize
def iswhole(n1,n2):
    return np.where(abs(n1-n2)<1e-4,True,False)
```

```python
#Non-Linearity Function
def NonLinearity(x,non_linearity='ReLU',a=None):
    if non_linearity == 'sigmoid':
        return 1/(1+np.exp(-1*x))
    elif non_linearity == 'ReLU':
        x[x<0]=0
        return x
    elif non_linearity == 'tanh':
        return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    elif non_linearity == 'PReLU':
        if a is None:
            a = np.random.rand(1)
        y=np.where(x > 0, x, x * a)
        return y
```

# Q1

```python
#Helper function for Convolution. Given image(after padding) and kernel, 
#it does convolution
def _convolution(image,kernel,stride=2):
   
    H,W=image.shape[0],image.shape[1]
    h,w=kernel.shape[0],kernel.shape[1]
    
    hf,wf=((H-h)//stride)+1,((W-w)//stride)+1    
    
    ans = np.zeros((hf,wf))
    
    for i in range(0,hf,stride):
        for j in range(0,wf,stride):
            ans[i,j] = np.sum(image[i:i+kernel.shape[0],j:j+kernel.shape[1]]*kernel)
    
    return ans
```

```python
#It does padding of given image, does convolution with help of 
#helper function.Thereafter applies specified non-linearity.
def convolution(image,kernel=gkern(),stride=1,p=0,non_linearity='ReLU',a=0.1):
    
    kernel = np.dstack([kernel] * image.shape[2])
    image2=np.zeros((image.shape[0]+2*p,image.shape[1]+2*p,image.shape[2]))
    image2[p:p+image.shape[0],p:p+image.shape[1]] = image
    x = _convolution(image2,kernel,stride=stride) 
   
    return NonLinearity(x,non_linearity,a)
```

```python
#Pooling Function
def pooling(image,mode='max',dim=[3,3],stride=3):
    
    image = image.astype(float)
    
    if mode=='global_avg':
        return np.mean(np.mean(image,axis=0),axis=0)
    
    H,W = image.shape[:2]
    h,w = dim
    H=float(H)
    W=float(W)
    h=float(h)
    w=float(w)
    
    hf,wf= (((H-h)//stride)+1,((W-w)//stride)+1)
    hf = int(hf)
    wf = int(wf)
    ans = np.zeros((hf,wf))
    
    if mode=='max':
        for i in range(0,hf,stride):
            for j in range(0,wf,stride):
                ans[i,j] = image[i:i+dim[0],j:j+dim[1]].max(axis=0).max(axis=0)
    elif mode=='mean':
        for i in range(0,hf,stride):
            for j in range(0,wf,stride):
                ans[i,j] = np.mean(np.mean(image[i:i+dim[0],j:j+dim[1]],axis=0),axis=0)
    return ans
```

```python
#Testing Pooling Function
modes = ['max','mean','global_avg']
j=1
plt.figure(figsize=(5,8))
for i in range(len(modes)):
    plt.subplot(3,2,j)
    plt.imshow(ans2)
    plt.title('Input')
    
    j=j+1
    
    plt.subplot(3,2,j)
    pooled=pooling(ans2,mode=modes[i])
    plt.title('Output')
    try:
        plt.imshow(pooled)
    except TypeError:
        print("Cannot print single value,Global Average:")
        print(pooled)
    j+j+1
```

```python
#Convolution Layer Function
def ConvLayer(image,filters=[own,gkern(),mkern()],stride=1,p=0,non_linearity='ReLU',show=True):
    v=[]
    for j in range(len(filters)):
        v.append(convolution(image,filters[j],stride,p,non_linearity))
     
    if show==True:
        j=1
        plt.figure(figsize=(8,9))
        for i in range(len(filters)):
            for j in range(1+3*i,4+3*i):
                if j%3==1:
                    plt.subplot(len(filters),3,j)
                    plt.title('Image')
                    plt.imshow(image)
                
                elif j%3==2:
                    plt.subplot(len(filters),3,j)
                    plt.title('Kernel')
                    plt.imshow(np.dstack([filters[i]] * image.shape[2]))

                else:
                    plt.subplot(len(filters),3,j)
                    plt.title('Output') 
                    plt.imshow(v[i])
                    break
    return v
```

```python
#Pooling Layer Function
def PoolingLayer(convolved_dataset,mode='mean',dim=[3,3],stride=3):
    v=[]
    for j in range(len(convolved_dataset)):
            v.append(pooling(np.array(convolved_dataset[j]),mode,dim,stride))
    for i in range(len(convolved_dataset)):
        for j in range(1+2*i,3+2*i):
                if j%2==1:
                    plt.subplot(len(convolved_dataset),2,j)
                    plt.title('Image')
                    plt.imshow(convolved_dataset[i])
                    
                else:
                    plt.subplot(len(convolved_dataset),2,j)
                    plt.title('Output') 
                    plt.imshow(v[i])
                    break
        
    return np.array(v)
```

```python
#Helper Function for MLP
def ForwardPass(inputvector,insize=None,outsize=None,bias=None,arch=None):
    if arch is None:
        arch = np.random.rand(outsize,insize)
    if bias is None:
        bias = np.random.rand(outsize)
        
    try:
        output=arch*inputvector
    except ValueError:
        output=arch.T*inputvector
    output=np.sum(output,axis=1)
    output = output + bias
    return output
```

```python
def MLP(size,inputvector=None,nonlinearity='ReLU',hasSoftmax=True,architecture=None):
    if architecture is None:
        architecture = [np.random.rand(inputvector.shape[0],size[0])]
        for i in range(1,len(size)):
            architecture.append(np.random.rand(size[i-1],size[i]))
        #architecture = np.array(architecture,dtype=np.ndarray)
        bias = [np.random.rand(size[j]) for j in range(len(size))]
    answer = inputvector
    for i in range(len(architecture)):
        answer = ForwardPass(answer,bias=bias[i],arch=architecture[i])
        answer = NonLinearity(answer,non_linearity=nonlinearity)
        
    if hasSoftmax:
        val = np.sum(answer)
        answer = np.exp(answer)
        answer = answer/np.sum(answer)
        
    return answer
```

```python
#Helper function for Neural Network
def getShape(image,bunchoffilters=None,stride=1,dim=None):
    if bunchoffilters is not None:
        hf=int((image.shape[0]-bunchoffilters.shape[1])/stride)+1
        wf=int((image.shape[1]-bunchoffilters.shape[2])/stride)+1
        df=int(bunchoffilters.shape[0])
        return (hf,wf,df)
    else:
        hf=int((image.shape[0]-dim[0])/stride)+1
        wf=int((image.shape[1]-dim[1])/stride)+1
        df=int(image.shape[2])
        return (hf,wf,df)       
```

```python
#Amalgamation of all Helper Functions for Original Question
def FF(image,bunchoffilters1,bunchoffilters2,pl1,pl2,s1,s2):
    
    #First Step of Convolution layer with 16 kernels of size 3 × 3
    firststep=np.empty(getShape(image,bunchoffilters=bunchoffilters1), dtype='int')
    
    for i in range(bunchoffilters1.shape[0]):
          firststep[:,:,i]=convolution(image,kernel=bunchoffilters1[i],non_linearity='sigmoid')
    
    #Second Step of Max pooling layer of size 2 × 2 with a stride of 2
    secondstep=np.empty(getShape(firststep,stride=s1,dim=pl1), dtype='int')

    for i in range(firststep.shape[-1]):
        secondstep[:,:,i]=pooling(firststep[:,:,i],mode='max',dim=pl1,stride=s1)

    
    #Third Step of Convolution layer with 8 kernels of size 3 × 3
    thirdstep=np.empty(getShape(secondstep,bunchoffilters=bunchoffilters2), dtype='int')
    
    for i in range(bunchoffilters2.shape[0]):
        thirdstep[:,:,i]=convolution(secondstep,kernel=bunchoffilters2[i],non_linearity='sigmoid')
    
    
    #Fourth Step of Max pooling layer of size 2 × 2 with a stride of 2
    fourthstep=np.empty(getShape(thirdstep,stride=s2,dim=pl2), dtype='int')

    for i in range(thirdstep.shape[-1]):
        fourthstep[:,:,i]=pooling(thirdstep[:,:,i],mode='max',dim=pl2,stride=s2)
    
    
    #Fifth Step of Global Average Pooling 
    fifthstep=pooling(fourthstep,'global_avg')

    
    #Sixth Step of MLP with one hidden layer    
    finaloutput=MLP([fifthstep.shape[0],10],inputvector=fifthstep,nonlinearity='sigmoid')

    
    return np.array(fifthstep),finaloutput
```
